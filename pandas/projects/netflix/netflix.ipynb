## Task 2: What year did netflix add the most content to its platform

```python
data
```

```python
data['date_added'] = data['date_added'].str.strip()  # remove the leading or any extra space
data['new_date'] = pd.to_datetime(data['date_added'])
data['year'] = data['new_date'].dt.year
```

```python
# data['year'] = data['year'].fillna(0).astype(int)
```

```python
data['year']
```

```python
data['count'] = 1
```

```python
data_added_years_summed = data.groupby(['year'])['count'].sum().reset_index()
```

```python
output = data_added_years_summed.sort_values(['count'], ascending = False)
output['year'] = output['year'].astype(int)
output
```

```python
years = []
count = []
for index, row in output.iterrows():
    years.append(row['year'])
    count.append(row['count'])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(output['year'], output['count'])
plt.xticks(years, rotation = 'vertical')
plt.show()
```

## Task 5: Which actor/actress shows up most frequently in the netflix database

```python
pd.options.display.max_colwidth = 150
```

```python
data[['title', 'cast']]
```

```python
title_and_cast_df = data[['title', 'cast']]
title_and_cast_df = title_and_cast_df.dropna()
```

```python
cast_names_counter = {}
for index, row in title_and_cast_df.iterrows():
    movie_cast = row['cast']
    movie_cast_split = movie_cast.split(',')
    movie_cast_stripped = [name.strip() for name in movie_cast_split]
    
    for name in movie_cast_stripped:
        cast_names_counter[name] = cast_names_counter.get(name, 0) + 1
        
# print(max(cast_names_counter, key=cast_names_counter.get))
#Anupam Kher

dict(sorted(cast_names_counter.items(), key = lambda item:item[1], reverse = True))
```
